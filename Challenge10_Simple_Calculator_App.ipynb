**Built-in Functions and User-Defined Functions**

**Project Title:** Simple Calculator App

**Scenario:**

 A company wants to provide users with a simple calculator to perform basic mathematical operations.

**Tasks:**

- Create functions for addition, subtraction, multiplication, and division.
- Ask the user to input two numbers and the operation they want to perform.
- Call the appropriate function and display the result.
- Use built-in functions like abs(), round(), and pow() within your solution.
- Handle division by zero gracefully and inform the user.

```python
# 1. User-defined calculator functions

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error: Cannot divide by zero!"
    return a / b

# 2. User Input Section

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

print("\nChoose Operation:")
print("1. Addition (+)")
print("2. Subtraction (-)")
print("3. Multiplication (*)")
print("4. Division (/)")
choice = input("Enter your choice (1-4): ")

# 3. Perform operation based on user choice

if choice == "1":
    result = add(num1, num2)

elif choice == "2":
    result = subtract(num1, num2)

elif choice == "3":
    result = multiply(num1, num2)

elif choice == "4":
    result = divide(num1, num2)

else:
    result = "Invalid choice!"

# 4. Use Built-in Functions

print("\n--- Result Summary ---")

# pow() â€“ example: square of result (if result is numeric)
if isinstance(result, (int, float)):
    print("Result:", round(result, 2))          # rounded output
    print("Absolute Value:", abs(result))       # absolute value
    print("Squared (pow):", pow(result, 2))     # power function
else:
    print(result)  # print error message

```
